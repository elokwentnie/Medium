```python
# Import necessary libraries
import pandas as pd
from sklearn.datasets import load_wine

# Load wine dataset from sklearn library
wine_data = load_wine()

# Create first dataframe with wine information 
wine_info_df = pd.DataFrame(data=wine_data.data, columns=wine_data.feature_names) 

# Create second dataframe with target information (will be used and explained later)
wine_target_df = pd.DataFrame({'target': wine_data.target})

#Let's look at our dataframe
wine_info_df
```

```python
# Check what are the names of columns in the dataset
print(wine_info_df.columns)

# Drop 'ash' column and create new DataFrame with the same name as before
wine_info_df = wine_info_df.drop(columns=['ash'])

# Chech whether the column is still in the dataset
print('ash' in wine_info_df.columns)
```

```python
# Drop 'malic_acid'
wine_info_df.drop(columns=['malic_acid'])
# Data will be displayed without `malic_acid' column
```

```python
# Chech whether the 'malic_acid' column is still in the dataset
print('malic_acid' in wine_info_df.columns)
```

```python
# To avoid that, we can just use `inplace=True`
wine_info_df.drop(columns=['malic_acid'], inplace=True)
```

```python
# Chech whether the 'malic_acid' column is still in the dataset
print('malic_acid' in wine_info_df.columns)
# Finally it worked :) 
```

```python
# Renaming column to add more context
wine_info_df = wine_info_df.rename(columns={'alcohol': 'alcohol [%]'})
```

```python
# Merge two DataFrames together
merged_df = pd.merge(wine_info_df, wine_target_df, left_index=True, right_index=True)
merged_df
```

We can see new column 'target'

```python
# Concatenate DataFrames along columns, if dimensions are not the same, missing values will be replaced with 'NaN'
additional_info_df = pd.DataFrame({'additional_info': ['Info_A', 'Info_B', 'Info_C']})
concatenated_df = pd.concat([wine_info_df, additional_info_df], axis=1)
concatenated_df
```

```python
# Group data by target and aggregate alcohol by its average for given target and maximum value for proline
grouped_data = merged_df.groupby('target').agg({'alcohol [%]': 'mean', 'proline': 'max'})
grouped_data
```
