````python
# 'education-num' represents the numerical encoding of the education level
# it is an alternative of the 'education' column
# we can rename it to 'education-level' to make it the column name more understandable
````

````python
adult_df = adult_df.rename(columns={'education_num': 'education-level'})
````

````python
for col in adult_df.columns:
    print(col)
````

````python
data_education_info = {'education': ['Bachelors', 'Masters', 'Doctorate'],
                        'degree_type': ['BA', 'MA', 'PhD']}
df_education_info = pd.DataFrame(data_education_info)

# Merge based on the 'education' column
merged_df = pd.merge(adult_df, df_education_info, on='education', how='left')

# Display the merged DataFrame
merged_df.head()
````

* `merge` is used to combine two or more DataFrames based on the values in specified columns or indivies, you can also specify what kind of join should be used, adding the `how` parameter
* Syntax: 
```
pd.merge(df1, df2, on='common_column', how='left')
```

````python
# Concatenate adult_df with itself
concatenated_df = pd.concat([adult_df, adult_df], axis=0)

# Display the first few rows of the concatenated DataFrame
concatenated_df.info()
````

* `concatenate` is used to concatenate data along a particular axis (rows or columns), regardless of the values in the DataFrames.
* Syntax: 
```
pd.concat([df1, df2], axis=0) #along rows
pd.concat([df1, df2], axis=1) #along columns
```

````python
# Group data by education-level and aggregate age and hours_per_week by its median 
grouped_data = adult_df.groupby('education-level').agg({'age': 'median', 'hours_per_week': 'median'})
grouped_data
````

````python
adult_df['education'] = adult_df['education'].astype('category')
adult_df['education'].info()
````

````python
adult_df
````

````python
# Create a new column 'income_category' based on income conditions
adult_df['income_category'] = adult_df['income'].apply(lambda x: 'High' if x == '>50K' else 'Low')
````

````python
# Binning 'age' into custom intervals
bins = [0, 25, 50, 100]
labels = ['Young', 'Adult', 'Senior']
adult_df['age_group'] = pd.cut(adult_df['age'], bins=bins, labels=labels)
````

````python
# Replace 'sex' values with numerical representation
gender_mapping = {'Male': 0, 'Female': 1}
adult_df['sex_numeric'] = adult_df['sex'].replace(gender_mapping)
adult_df
````

````python
# Create a pivot table for average hours per week for each education level
pivot_table_education = adult_df.pivot_table(index='education', values=['hours_per_week', 'capital_gain'], aggfunc='mean')
pivot_table_education
````
