```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Generating example data
y_true = np.array([1, 0, 1, 1, 0, 1, 0, 1, 0, 1])  # True labels
y_pred = np.array([1, 1, 0, 1, 0, 1, 1, 0, 1, 1])  # Predicted labels

# Creating confusion matrix
cm = confusion_matrix(y_true, y_pred)

# Calculating accuracy, precision, and recall
accuracy = np.trace(cm) / np.sum(cm)
precision = cm[1, 1] / np.sum(cm[:, 1])
recall = cm[1, 1] / np.sum(cm[1, :])

# Switching to xkcd style
with plt.xkcd():
    # Visualizing confusion matrix
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False, square=True)
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.title(f"Confusion Matrix\nAccuracy: {accuracy:.2f}, Precision: {precision:.2f}, Recall: {recall:.2f}")

    # Adding xkcd style sketchiness
    plt.xkcd()

    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example values for accuracy, precision, and recall
accuracy = 0.85
precision = 0.78
recall = 0.92

# Labels for the bars
labels = ['Accuracy', 'Precision', 'Recall']

# Values for the bars
values = [accuracy, precision, recall]

# Plotting the bar chart
plt.bar(labels, values, color=['blue', 'green', 'orange'])

# Adding text labels with the corresponding values
for label, value in zip(labels, values):
    plt.text(label, value + 0.02, f'{value:.2f}', ha='center', va='bottom')

# Adding a title
plt.title('Accuracy, Precision, and Recall')

# Display the plot
plt.show()
```
